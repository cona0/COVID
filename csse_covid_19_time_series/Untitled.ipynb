```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
```

```python
datasets = {"Confirmed": pd.read_csv('time_series_19-covid-Confirmed.csv', header = 0)
           ,"Dead": pd.read_csv('time_series_19-covid-Deaths.csv', header = 0)
           ,"Recovered": pd.read_csv( 'time_series_19-covid-Recovered.csv', header = 0)}
```

```python
joined = dict()

for name, df in datasets.items():
    
    joined[name] = df.reset_index().drop(["index", "Lat", "Long"], axis = 1).set_index(["Country/Region", "Province/State"]).sum(level = "Country/Region")

sub = pd.concat(joined)
```

```python
shifts = [("Italy", 14), ("Spain", 7), ("United Kingdom", 0), ("China", 28)]
```

```python
countries = [
    "China",
    "France",
    "United Kingdom",
    "Italy"
]
```

```python
fig, ax = plt.subplots(figsize = (15,10))
ax.set_xlabel("Date")
ax.set_ylabel("Cases")

for country, shift in shifts:
    
    sub.loc(axis = 0)["Confirmed", country].T.shift(shift).plot(logy = True , ax = ax)

ax.legend()
```
